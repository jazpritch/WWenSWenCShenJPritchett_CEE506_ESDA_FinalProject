```python
import os
import requests
from datetime import datetime, timedelta

# Base URL for NCSS requests
base_url = "https://apps.glerl.noaa.gov/thredds/ncss/grid/glsea_nc/{year}/{month}/{filename}?var=sst&north=42.5&west=-83.6&east=-78.9&south=41.3&horizStride=1&time_start={date}T12:00:00Z&time_end={date}T12:00:00Z&&&accept=netcdf4-classic"

# Output directory to save downloaded files
output_dir = "Lake_Erie_Daily_Data"
os.makedirs(output_dir, exist_ok=True)

# Loop through years, months, and days
start_date = datetime(2009, 1, 1)
end_date = datetime(2023, 12, 31)
current_date = start_date

while current_date <= end_date:
    year = current_date.year
    month = f"{current_date.month:02d}"
    day_of_year = (current_date - datetime(current_date.year, 1, 1)).days + 1
    filename = f"{year}_{day_of_year:03d}_glsea_sst.nc"
    
    # Construct the full URL
    request_url = base_url.format(
        year=year,
        month=month,
        filename=filename,
        date=current_date.strftime("%Y-%m-%d")
    )
    
    # Local file path
    local_file = os.path.join(output_dir, f"{year}_{current_date.strftime('%m%d')}.nc")
    
    # Check if the file already exists
    if os.path.exists(local_file):
        print(f"File already exists: {local_file}")
    else:
        print(f"Downloading {local_file} from {request_url}")
        try:
            response = requests.get(request_url)
            response.raise_for_status()  # Check for request errors
            with open(local_file, 'wb') as f:
                f.write(response.content)
            print(f"Saved: {local_file}")
        except requests.exceptions.RequestException as e:
            print(f"Failed to download {local_file}: {e}")
    
    # Move to the next day
    current_date += timedelta(days=1)
```

```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from datetime import datetime, timedelta


data_dir = "Lake_Erie_Daily_Data" 
file_pattern = os.path.join(data_dir, "*.nc")
files = sorted(glob.glob(file_pattern))


print(f"Number of NetCDF files: {len(files)}")


all_dates = []
sst_means = []


start_date = datetime(2009, 1, 1)


for i, file in enumerate(files):
    with Dataset(file, 'r') as dataset:
        # Extract SST and replace fill values with NaN
        sst = dataset.variables['sst'][:]
        sst_cleaned = np.where(sst == -99999.0, np.nan, sst)
        
   
        mean_sst = np.nanmean(sst_cleaned)
        sst_means.append(mean_sst)
        
      
        all_dates.append(start_date + timedelta(days=i))


all_dates = np.array(all_dates)
sst_means = np.array(sst_means)


print(f"First 10 Dates: {all_dates[:10]}")
print(f"First 10 Mean SST Values: {sst_means[:10]}")
print(f"Number of Dates: {len(all_dates)}")
print(f"Number of SST Mean Values: {len(sst_means)}")


nan_count = np.sum(np.isnan(sst_means))
print(f"Number of NaN values in SST means: {nan_count}")


if nan_count > 0:
    print("Replacing NaN values with linear interpolation for smoother plotting.")
    valid_indices = ~np.isnan(sst_means)
    sst_means = np.interp(np.arange(len(sst_means)), np.where(valid_indices)[0], sst_means[valid_indices])


plt.figure(figsize=(12, 6))
plt.plot(all_dates, sst_means, label="Mean SST", color="blue")
plt.title("Lake Erie Average Surface Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Average Surface Temperature (C)")
plt.legend()
plt.grid(True)
plt.gcf().autofmt_xdate()  
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os
import glob
from datetime import datetime, timedelta
from multiprocessing import Pool, cpu_count

# Directory containing NetCDF files
data_dir = "Lake_Erie_Daily_Data"
files = sorted(glob.glob(f"{data_dir}/*.nc"))

# Load longitude and latitude from the first file
with nc.Dataset(files[0], 'r') as dataset:
    lons = dataset.variables['lon'][:]
    lats = dataset.variables['lat'][:]

# Define seasons and months
seasons = {
    "Winter": [12, 1, 2],
    "Spring": [3, 4, 5],
    "Summer": [6, 7, 8],
    "Fall": [9, 10, 11]
}

# Function to process a single file
def process_file(file, year, months):
    try:
        with nc.Dataset(file, 'r') as dataset:
            # Extract the date from the file
            time_var = dataset.variables['time']
            base_date = datetime(1970, 1, 1)  # Assuming "seconds since 1970-01-01"
            file_date = base_date + timedelta(seconds=int(time_var[:][0]))

            # Check if the file belongs to the current season and year
            if file_date.year == year and file_date.month in months:
                sst = dataset.variables['sst'][:].data
                sst[sst == -99999.0] = np.nan  # Replace fill values with NaN
                return sst
    except Exception as e:
        print(f"Error processing {file}: {e}")
    return None

# Function to process data for a single season and year
def process_season(year_season):
    year, season, months = year_season
    print(f"Processing {season} {year}...")
    seasonal_sst = []
    for file in files:
        sst = process_file(file, year, months)
        if sst is not None:
            seasonal_sst.append(sst)

    if seasonal_sst:
        seasonal_avg = np.nanmean(np.array(seasonal_sst), axis=0)
        return year, season, seasonal_avg
    return year, season, None

# Prepare year-season combinations
year_season_combinations = [
    (year, season, months)
    for year in range(2009, 2024)
    for season, months in seasons.items()
]

# Parallel processing
with Pool(processes=cpu_count()) as pool:
    results = pool.map(process_season, year_season_combinations)

# Save results and generate plots
for year, season, sst_avg in results:
    if sst_avg is not None:
        plt.figure(figsize=(10, 6))
        plt.imshow(np.squeeze(sst_avg), origin='lower', cmap='viridis', extent=[
            lons.min(), lons.max(), lats.min(), lats.max()
        ])
        plt.colorbar(label="Surface Temperature (C)")
        plt.title(f"Lake Erie {season} Average Surface Temperature - {year}")
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.savefig(f"lake_erie_{year}_{season.lower()}_sst.png")
        plt.close()

print("Processing complete!")
```

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob

data_dir = "Lake_Erie_Daily_Data"
files = sorted(glob.glob(f"{data_dir}/*.nc"))

ds = xr.open_mfdataset(files, combine="by_coords", parallel=True, decode_times=False)
all_dates = pd.to_datetime(ds["time"].values, origin='1970-01-01', unit='s')
sst_cleaned = ds["sst"].where(ds["sst"] != -99999.0).mean(dim=["lat", "lon"]).values

if len(all_dates) != len(sst_cleaned):
    raise ValueError("Mismatch between dates and SST data length!")

data = pd.DataFrame({'Date': all_dates, 'Temperature': sst_cleaned})
data.set_index('Date', inplace=True)

weekly_stats = data.resample('W').mean()
monthly_stats = data.resample('M').mean()

seasons = {"Winter": [12, 1, 2], "Spring": [3, 4, 5], "Summer": [6, 7, 8], "Fall": [9, 10, 11]}
seasonal_data = {
    season: data[data.index.month.isin(months)].resample('Y').mean()
    for season, months in seasons.items()
}

weekly_stats.plot(figsize=(12, 6), title="Weekly Average SST (2009-2023)")
plt.ylabel("Temperature (C)")
plt.xlabel("Date")
plt.grid()
plt.savefig("weekly_avg_sst.png")

monthly_stats.plot(figsize=(12, 6), title="Monthly Average SST (2009-2023)")
plt.ylabel("Temperature (C)")
plt.xlabel("Date")
plt.grid()
plt.savefig("monthly_avg_sst.png")

plt.figure(figsize=(12, 6))
for season, data in seasonal_data.items():
    plt.plot(data.index, data["Temperature"], label=season)
plt.legend()
plt.title("Seasonal Average SST (2009-2023)")
plt.ylabel("Temperature (C)")
plt.xlabel("Year")
plt.grid()
plt.savefig("seasonal_avg_sst.png")
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from pykrige.ok import OrdinaryKriging
from shapely.geometry import Polygon, MultiPolygon

# Load the data
lake_erie_boundary = gpd.read_file("lake-erie_812.geojson")
data = pd.read_csv("cleaned_water_temp.csv")  # Replace with your file name
data["datetime"] = pd.to_datetime(data["ActivityStartDate"])  # Ensure correct datetime parsing
gdf = gpd.GeoDataFrame(
    data,
    geometry=gpd.points_from_xy(data["ActivityLocation/LongitudeMeasure"], data["ActivityLocation/LatitudeMeasure"]),
    crs=lake_erie_boundary.crs,
)

# Define grid for interpolation
grid_lon, grid_lat = np.meshgrid(
    np.linspace(-83.6, -78.9, 200),
    np.linspace(41.3, 42.8, 200),
)

# Extract boundary points
boundary_points = []
for geom in lake_erie_boundary.geometry:
    if isinstance(geom, Polygon):
        boundary_points.extend(np.array(geom.exterior.coords))
    elif isinstance(geom, MultiPolygon):
        for part in geom.geoms:
            boundary_points.extend(np.array(part.exterior.coords))
boundary_points = np.array(boundary_points)

# Add pseudo-points along the boundary
pseudo_points = pd.DataFrame({
    "longitude": boundary_points[:, 0],
    "latitude": boundary_points[:, 1],
    "WaterTemperature": np.mean(data["ResultMeasureValue"]),  # Replace with column for water temperature
})
pseudo_gdf = gpd.GeoDataFrame(
    pseudo_points, geometry=gpd.points_from_xy(pseudo_points["longitude"], pseudo_points["latitude"]), crs=gdf.crs
)

# Seasonal months
seasonal_months = {
    "Winter": [12, 1, 2],
    "Spring": [3, 4, 5],
    "Summer": [6, 7, 8],
    "Fall": [9, 10, 11],
}

# Process data for each year and season
def process_season(year, season, months):
    subset = gdf[(gdf.datetime.dt.year == year) & (gdf.datetime.dt.month.isin(months))]
    if len(subset) > 10:  # Ensure sufficient data points for kriging
        combined_gdf = pd.concat([subset, pseudo_gdf], ignore_index=True)

        # Ordinary Kriging
        kriging = OrdinaryKriging(
            combined_gdf["longitude"], combined_gdf["latitude"], combined_gdf["ResultMeasureValue"],
            variogram_model="linear",
        )
        z, ss = kriging.execute("grid", grid_lon[0], grid_lat[:, 0])

        # Masking kriging results to lake boundary
        grid_gdf = gpd.GeoDataFrame(
            {"geometry": gpd.points_from_xy(grid_lon.flatten(), grid_lat.flatten())},
            crs=lake_erie_boundary.crs,
        )
        grid_gdf["value"] = z.flatten()
        grid_within = gpd.sjoin(grid_gdf, lake_erie_boundary, how="inner", predicate="within")

        z_masked = np.full(z.shape, np.nan)
        for idx, row in grid_within.iterrows():
            lon_idx = np.argmin(np.abs(grid_lon[0] - row.geometry.x))
            lat_idx = np.argmin(np.abs(grid_lat[:, 0] - row.geometry.y))
            z_masked[lat_idx, lon_idx] = row["value"]

        # Plotting
        plt.figure(figsize=(10, 6))
        plt.imshow(
            z_masked, origin="lower", cmap="viridis",
            extent=[grid_lon.min(), grid_lon.max(), grid_lat.min(), grid_lat.max()],
        )
        lake_erie_boundary.boundary.plot(ax=plt.gca(), color="black", linewidth=1, label="Lake Boundary")
        plt.colorbar(label="Water Temperature (C)")
        plt.title(f"Lake Erie {season} Water Temperature - {year}")
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.legend()
        plt.savefig(f"water_temp_{year}_{season}.png")
        plt.close()

# Process all years and seasons
def process_years(start_year, end_year):
    for year in range(start_year, end_year + 1):
        print(f"Processing year {year}...")
        for season, months in seasonal_months.items():
            process_season(year, season, months)
        print(f"Finished year {year}.")

process_years(2009, 2023)
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from pykrige.ok import OrdinaryKriging
from shapely.geometry import Polygon, MultiPolygon

# Load Lake Erie boundary and temperature data
lake_erie_boundary = gpd.read_file("lake-erie_812.geojson")
data = pd.read_csv("cleaned_water_temp.csv")
data["datetime"] = pd.to_datetime(data["datetime"])  # Correct column for datetime
gdf = gpd.GeoDataFrame(
    data,
    geometry=gpd.points_from_xy(data["longitude"], data["latitude"]),
    crs=lake_erie_boundary.crs
)

# Define grid for kriging
grid_lon, grid_lat = np.meshgrid(
    np.linspace(-83.6, -78.9, 200),
    np.linspace(41.3, 42.8, 200)
)

# Extract boundary points for pseudo-points
boundary_points = []
for geom in lake_erie_boundary.geometry:
    if isinstance(geom, Polygon):
        boundary_points.extend(np.array(geom.exterior.coords))
    elif isinstance(geom, MultiPolygon):
        for part in geom.geoms:
            boundary_points.extend(np.array(part.exterior.coords))
boundary_points = np.array(boundary_points)

# Generate pseudo-points
pseudo_points = pd.DataFrame({
    "longitude": boundary_points[:, 0],
    "latitude": boundary_points[:, 1],
    "temperature_mean": np.nanmean(data["temperature_mean"]),  # Using average mean temperature
})
pseudo_gdf = gpd.GeoDataFrame(
    pseudo_points,
    geometry=gpd.points_from_xy(pseudo_points["longitude"], pseudo_points["latitude"]),
    crs=gdf.crs
)

def process_season(year, season, months):
    subset = gdf[(gdf["datetime"].dt.year == year) & (gdf["datetime"].dt.month.isin(months))]
    if not subset.empty:
        combined_gdf = pd.concat([subset, pseudo_gdf], ignore_index=True)

        # Ordinary Kriging
        kriging = OrdinaryKriging(
            combined_gdf["longitude"], combined_gdf["latitude"], combined_gdf["temperature_mean"],
            variogram_model="linear"
        )
        z, ss = kriging.execute("grid", grid_lon[0], grid_lat[:, 0])

        # Mask outside the boundary
        grid_gdf = gpd.GeoDataFrame(
            {"geometry": gpd.points_from_xy(grid_lon.flatten(), grid_lat.flatten())},
            crs=lake_erie_boundary.crs,
        )
        grid_gdf["value"] = z.flatten()
        grid_within = gpd.sjoin(grid_gdf, lake_erie_boundary, how="inner", predicate="within")

        z_masked = np.full(z.shape, np.nan)
        for idx, row in grid_within.iterrows():
            lon_idx = np.argmin(np.abs(grid_lon[0] - row.geometry.x))
            lat_idx = np.argmin(np.abs(grid_lat[:, 0] - row.geometry.y))
            z_masked[lat_idx, lon_idx] = row["value"]

        # Plot
        plt.figure(figsize=(10, 6))
        plt.imshow(
            z_masked, origin="lower", cmap="viridis",
            extent=[grid_lon.min(), grid_lon.max(), grid_lat.min(), grid_lat.max()]
        )
        lake_erie_boundary.boundary.plot(ax=plt.gca(), color="black", linewidth=1, label="Lake Boundary")
        plt.colorbar(label="Mean Water Temperature (C)")
        plt.title(f"Lake Erie {season} Water Temperature - {year}")
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.legend()
        plt.savefig(f"water_temp_{year}_{season}.png")
        plt.close()
    else:
        print(f"Not enough data for {year} {season}. Skipping.")

def process_year(year):
    seasonal_months = {
        "Winter": [12, 1, 2],
        "Spring": [3, 4, 5],
        "Summer": [6, 7, 8],
        "Fall": [9, 10, 11]
    }
    for season, months in seasonal_months.items():
        process_season(year, season, months)
    print(f"Finished processing {year}.")

# Process years 2009-2023
for year in range(2009, 2024):
    process_year(year)
```
