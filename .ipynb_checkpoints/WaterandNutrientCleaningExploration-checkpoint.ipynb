Nutrient Data 

```python
# Load and process nutrient data
def process_nutrient_data(file_path):
    nutrient_data = pd.read_csv(file_path)
    nutrient_data['date'] = pd.to_datetime(nutrient_data['storm_start'])
    nutrient_columns = [
        'USGS_Station_Number', 'date', 'nitrate_plus_nitrite_conc_mgL',
        'orthophosphate_conc_mgL', 'total_phosphorus_unfiltered_conc_mgL'
    ]
    nutrient_data = nutrient_data[nutrient_columns]
    return nutrient_data

# Load the data
water_temp_data = process_water_temp_data('data/water_temperature/all_year_glsea_avg_e_F.csv')
nutrient_data = process_nutrient_data('data/nutrient_data/All_EOF_StormEventLoadsFormatted.csv')

# Add coordinates to nutrient data
nutrient_data['Latitude'] = nutrient_data['USGS_Station_Number'].map(lambda x: usgs_coordinates.get(x, (np.nan, np.nan))[0])
nutrient_data['Longitude'] = nutrient_data['USGS_Station_Number'].map(lambda x: usgs_coordinates.get(x, (np.nan, np.nan))[1])

# Remove rows with missing coordinates
nutrient_data = nutrient_data.dropna(subset=['Latitude', 'Longitude'])

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(
    nutrient_data, 
    geometry=gpd.points_from_xy(nutrient_data.Longitude, nutrient_data.Latitude),
    crs="EPSG:4326"
)

# Calculate monthly averages
monthly_temp = water_temp_data.set_index('date').resample('M').mean().reset_index()
monthly_nutrient = gdf.set_index('date').resample('M').mean().reset_index()

# Merge temperature and nutrient data
combined_data = pd.merge(monthly_temp, monthly_nutrient, on='date', how='outer')

# Plot temperature and nitrate concentration over time
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (F)', color='tab:blue')
ax1.plot(combined_data['date'], combined_data['temperature'], label='Temperature', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Nitrate Concentration (mg/L)', color='tab:green')
ax2.plot(combined_data['date'], combined_data['nitrate_plus_nitrite_conc_mgL'], label='Nitrate Concentration', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')

plt.title('Temperature and Nitrate Concentration Over Time')
fig.tight_layout()
plt.show()
```

```python
# Spatial analysis
# Calculate average nutrient concentrations for each station
station_avg = gdf.groupby('USGS_Station_Number').agg({
    'nitrate_plus_nitrite_conc_mgL': 'mean',
    'orthophosphate_conc_mgL': 'mean',
    'total_phosphorus_unfiltered_conc_mgL': 'mean',
    'Latitude': 'first',
    'Longitude': 'first'
}).reset_index()

# Create a GeoDataFrame for the average concentrations
gdf_avg = gpd.GeoDataFrame(
    station_avg, 
    geometry=gpd.points_from_xy(station_avg.Longitude, station_avg.Latitude),
    crs="EPSG:4326"
)

# Plot the average nutrient concentrations
fig, ax = plt.subplots(figsize=(12, 8))
gdf_avg.plot(ax=ax, column='nitrate_plus_nitrite_conc_mgL', cmap='YlOrRd', legend=True, 
             legend_kwds={'label': 'Average Nitrate + Nitrite Concentration (mg/L)'})
ax.set_title('Average Nitrate + Nitrite Concentration by Station')
plt.show()

# Correlation analysis
corr_matrix = combined_data[['temperature', 'nitrate_plus_nitrite_conc_mgL', 
                             'orthophosphate_conc_mgL', 'total_phosphorus_unfiltered_conc_mgL']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Save processed data
combined_data.to_csv('processed_combined_data.csv', index=False)
gdf_avg.to_file('processed_nutrient_data_with_coords.geojson', driver='GeoJSON')

print("Data processing and analysis complete. Results saved to CSV and GeoJSON files.")
```

```python
def validate_station_coordinates(nutrient_data, usgs_coordinates):
    # Check for stations without coordinates
    stations_without_coords = set(nutrient_data['USGS_Station_Number']) - set(usgs_coordinates.keys())
    
    if stations_without_coords:
        print(f"Warning: {len(stations_without_coords)} stations lack coordinates:")
        print(stations_without_coords)
    else:
        print("All stations have coordinate mappings.")
    
    return len(stations_without_coords) == 0

# Run the validation
validate_station_coordinates(nutrient_data, usgs_coordinates)
```
